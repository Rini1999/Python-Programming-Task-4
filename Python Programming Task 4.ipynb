```python
#Question 5

def count_substring(string, sub_string):
    string_length = len(string)
    substring_length = len(sub_string)
    ctr = 0
    
    for i in range((string_length - substring_length) + 1):
        if(string[i:(i+substring_length)] == sub_string):
            ctr = ctr + 1
    return ctr

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
#Question 7

#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
#Question 10

def print_formatted(number):
    # your code goes here
    length = len("{0:b}".format(number))
    
    for i in range(1, number+1):
        print("{0:{w}d} {0:{w}o} {0:{w}X} {0:{w}b}".format(i, w=length))
        
if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
#Question 11

def print_rangoli(size):
    # your code goes here
    N = size
    lst = list(map(chr, range(97,123)))
    x = lst[N-1::-1] + lst[1:N]
    s = len('-'.join(x))
    for i in range(1,N):
        print('-'.join(lst[N-1:N-i:-1] + lst[N-i:N]).center(s,'-'))
    for i in range(N, 0, -1):
        print('-'.join(lst[N-1:N-i:-1] + lst[N-i:N]).center(s,'-'))
            
if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
